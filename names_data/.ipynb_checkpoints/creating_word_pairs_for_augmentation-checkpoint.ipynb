```python
import pandas as pd
import numpy as np
```

```python
names = pd.read_csv('names_race.csv')
```

```python
names
```

```python
names['pctasian'] = names['pctapi'] + names['pctaian'] 
```

```python
names
```

```python
names['name'] = names['firstname'].str.lower()
```

```python
names
```

```python
import gensim
from gensim.models import Word2Vec
 
model = Word2Vec.load("../wikipedia_corpus/wiki_en.txt.model")
```

```python
names_list = list(names['name'])
names_not_in_vocab = []
frequencies = {}

for name in names_list:
    try:
        vocab_obj = model.wv.vocab[name]
        frequencies[name] = vocab_obj.count
    except:
        names_not_in_vocab.append(name)
```

```python
frequencies
```

```python
freq = pd.DataFrame.from_dict(frequencies, orient='index', columns=['frequency'])
freq['name'] = freq.index
```

```python
freq
```

```python
full_names = names.merge(freq, how='left', on='name')
```

```python
full_names
```

```python
white = full_names[['name', 'pctwhite', 'frequency']]
hispanic = full_names[['name', 'pcthispanic', 'frequency']]
```

```python
white.sort_values(['pctwhite', 'frequency'], ascending=False)
```

```python
hispanic.sort_values(['pcthispanic', 'frequency'], ascending=False)
```

```python
surnames = pd.read_csv('census_surnames.csv')
```

```python
surname_list = list(surnames['NAME'].str.lower())
```

```python
surname_list
```

```python
surname_list = list(surnames['NAME'].str.lower())
surnames_not_in_vocab = []
frequencies_surnames = {}

for surname in surname_list:
    try:
        vocab_obj = model.wv.vocab[surname]
        frequencies_surnames[surname] = vocab_obj.count
    except:
        surnames_not_in_vocab.append(surname)
```

```python
frequencies_surnames
```

```python
surname_freq = pd.DataFrame.from_dict(frequencies_surnames, orient='index', columns=['frequency'])
surname_freq['surname'] = surname_freq.index
```

```python
surname_freq
```

```python
surnames['surname'] = surnames['NAME'].str.lower()
surnames = surnames.merge(surname_freq, how='left', on='surname')
surnames
```

```python
surnames.sort_values('[["PCTAPI"')
surnames['[["PCTAPI"'] = surnames['[["PCTAPI"'].str[2:-1]
```

```python
surnames['[["PCTAPI"'] = surnames['[["PCTAPI"'].replace('(S)', '0').astype(float)
surnames['PCTBLACK'] = surnames['PCTBLACK'].replace('(S)', '0').astype(float)
surnames['PCTAIAN'] = surnames['PCTAIAN'].replace('(S)', '0').astype(float)
surnames['PCTWHITE'] = surnames['PCTWHITE'].replace('(S)', '0').astype(float)
surnames['PCTHISPANIC'] = surnames['PCTHISPANIC'].replace('(S)', '0').astype(float)
surnames['PCTASIAN'] = surnames['PCTAIAN'] + surnames['[["PCTAPI"']
surnames = surnames.drop([ 'PCT2PRACE', 'NAME', 'RANK]', 'Unnamed: 9', 'PCTAIAN', '[["PCTAPI"'], axis=1)
```

```python
surnames = surnames.sort_values('frequency').dropna()
```
