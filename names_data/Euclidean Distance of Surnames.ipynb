```python
def get_surname_pairs(black, white, ptcblack):
    '''Function to get the surname pairs with min eucledian distance in the 
    frequency - race percentace space'''
    
    np_blackwhite = {}
    white_surnames_used = []

    for b in range(len(black)):
        black_surname = np.array((black['frequency'][b] ,black[ptcblack][b]))
        min_distance = float("inf")
        white_index = 0

        for w in range(len(white)):
            white_surname = np.array((white['frequency'][w] ,white['PCTWHITE'][w]))
            distance = np.linalg.norm(black_surname-white_surname)
            if distance < min_distance:
                if white['surname'][w] not in white_surnames_used:
                    min_distance = distance
                    white_index = w

        white_surnames_used.append(white['surname'][white_index])
        np_blackwhite[black['surname'][b]] = white['surname'][white_index]
        print(b)
        
    return np_blackwhite
```
